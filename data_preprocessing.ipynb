```python
import os
import cv2

DATADIR = "asl_alphabet_train/asl_alphabet_train/"
labels = [i for i in os.listdir("asl_alphabet_train/asl_alphabet_train/")]

DATA = []

for classes in labels:
    path = os.path.join(DATADIR, classes)
    #class_num = labels.index(classes)      # to save labels in int as index
    class_num = classes
    
    for i, image in enumerate(os.listdir(path)):
        # take only 50 samples from each class
        if i < 50:
            img = cv2.imread(os.path.join(path, image), 1)
            DATA.append([img, class_num])
        else:
            pass
```

```python
import cv2
import mediapipe as mp
import numpy as np

mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands

# For static images:
IMAGE_FILES = DATA
with mp_hands.Hands(
    static_image_mode=False,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:
  for image, label in IMAGE_FILES:
    # Read an image, flip it around y-axis for correct handedness output (see
    # above).
    #image = cv2.flip(cv2.imread(file), 1)
    # Convert the BGR image to RGB before processing.
    results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    # Print handedness and draw hand landmarks on the image.
    #print('Handedness:', results.multi_handedness)
    if not results.multi_hand_landmarks:
      continue
    image_height, image_width, _ = image.shape
    annotated_image = image.copy()
    for hand_landmarks in results.multi_hand_landmarks:
   
      fin_co = []   #finger co-ordinates
      for i, lm in enumerate(hand_landmarks.landmark):
        arr = [lm.x, lm.y]
        fin_co.append(arr)
      df = df.append({"0":fin_co[0][0], "0_1":fin_co[0][1], "1":fin_co[1][0], "1_1":fin_co[1][1], "2":fin_co[2][0], "2_1":fin_co[2][1], "3":fin_co[3][0], "3_1":fin_co[3][1],
                      "4":fin_co[4][0], "4_1":fin_co[4][1], "5":fin_co[5][0], "5_1":fin_co[5][1], "6":fin_co[6][0], "6_1":fin_co[6][1], "7":fin_co[7][0], "7_1":fin_co[7][1], 
                      "8":fin_co[8][0], "8_1":fin_co[8][1], "9":fin_co[9][0], "9_1":fin_co[9][1], "10":fin_co[10][0], "10_1":fin_co[10][1], "11":fin_co[11][0], "11_1":fin_co[11][1],
                      "12":fin_co[12][0], "12_1":fin_co[12][1], "13":fin_co[13][0], "13_1":fin_co[13][1], "14":fin_co[14][0], "14_1":fin_co[14][1], "15":fin_co[15][0], "15_1":fin_co[15][1],
                      "16":fin_co[16][0], "16_1":fin_co[16][1], "17":fin_co[17][0], "17_1":fin_co[17][1], "18":fin_co[18][0], "18_1":fin_co[18][1], "19":fin_co[19][0], "19_1":fin_co[19][1], 
                      "20":fin_co[20][0], "20_1":fin_co[20][1], "label":label}, ignore_index=True)
      
      #mask_label.append(label)
      
```
